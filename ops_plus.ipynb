# Calculating OPS+

The formula for OPS+ is:
```
    OPS/lgOPS * 100
```
However the problem is making the formula iterable across rows. I'm not sure about how to write a lambda, or function, that makes sure that Player A's OPS is divided by the lgOPS of that year.

````python
import advanced_statlines as stat 
import pandas as pd 
import numpy as np 
import mysql.connector as sql
from IPython import display

pd.options.display.max_columns = None
````

````python
batting = pd.read_csv('multiyear_batting_example.csv')
stat.format_df(batting)
batting = batting[batting.PA > 350]
batting = stat.order_by(batting, 'OPS', False)
batting.head()
````

````python
ops_plus = lambda row: row.OPS * 100 / batting.groupby(row.yearID).OPS.mean()
batting['OPS'] = batting.apply(ops_plus, axis=1)
````
